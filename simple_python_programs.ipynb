```python
def factorial(n):
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Get input from the user
number_str = input("Enter a non-negative integer to compute its factorial: ")
number = int(number_str) # This will raise a ValueError if input is not an integer

if number < 0:
    print("Factorial is not defined for negative numbers.")
else:
    print(f"The factorial of {number} is {factorial(number)}")
```

```python
# Get input from the user
number_str = input("Enter an integer to see its multiplication table: ")
number = int(number_str) # This will raise a ValueError if input is not an integer

print(f"\nMultiplication Table for {number}:")
for i in range(1, 11):  # Iterate from 1 to 10 inclusive
    product = number * i
    print(f"{number} x {i} = {product}")
```

```python
# Get input from the user
number_str = input("Enter an integer: ")
number = int(number_str) # This will raise a ValueError if input is not an integer

if number % 2 == 0:
    print(f"{number} is an even number.")
else:
    print(f"{number} is an odd number.")
```

```python
# Get input from the user
number_str = input("Enter a number to check if it's a palindrome: ")

# Convert the input string to an integer first (as it's a 'number')
# Then convert it back to a string for palindrome check
# Note: This will raise a ValueError if input is not an integer due to no try-except
original_number_str = str(int(number_str))

# Reverse the string
reversed_number_str = original_number_str[::-1]

# Check if it's a palindrome
if original_number_str == reversed_number_str:
    print(f"The number {number_str} is a palindrome.")
else:
    print(f"The number {number_str} is not a palindrome.")
```

## Simple Decision-Making System (No Libraries, No Try/Except)

Under the constraints of no external libraries and no `try-except` blocks, an "Expert System" for simple decision-making would primarily involve:

1.  **Facts:** Simple variables holding information.
2.  **Rules:** `if-elif-else` statements or functions that evaluate these facts to reach a conclusion.
3.  **Inference:** Direct execution of these rules based on the available facts.

This approach lacks the dynamic rule management, sophisticated inference engines, or robust error handling of a full-fledged expert system, but it can still make basic decisions based on predefined logic.

```python
# Define 'facts' (input from user)
print("--- Simple Decision-Making System ---")
print("Please answer the following questions to help make a decision.")

weather_input = input("What is the weather like today? (sunny/cloudy/rainy): ")
temperature_str = input("What is the temperature in Celsius? ")
temperature = int(temperature_str) # No try-except, so invalid input will raise ValueError

# Define a simple rule-based 'inference engine'
def make_activity_decision(weather, temp):
    if weather == "sunny" and temp >= 25:
        return "It's a great day for a picnic!"
    elif weather == "sunny" and temp < 25:
        return "A walk in the park would be nice."
    elif weather == "cloudy" and temp >= 15:
        return "Perhaps a museum visit or coffee with friends."
    elif weather == "rainy":
        return "Stay indoors and read a book or watch a movie."
    else:
        return "The weather is unusual, consider checking local events."

# Make a decision based on the facts
decision = make_activity_decision(weather_input, temperature)
print(f"\nBased on your input, the suggested activity is: {decision}")
```
