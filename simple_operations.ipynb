### Factorial Calculation

This code defines a function `factorial` that calculates the factorial of a non-negative integer using recursion. It then calls the function with an example number and prints the result.

```python
def factorial(n):
    """Calculates the factorial of a non-negative integer n."""
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
number = 5
result = factorial(number)
print(f"The factorial of {number} is {result}")
```
