### Factorial Calculation

This code defines a function `factorial` that calculates the factorial of a non-negative integer using recursion. It then calls the function with an example number and prints the result.

````python
def factorial(n):
    """Calculates the factorial of a non-negative integer n."""
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
number = 5
result = factorial(number)
print(f"The factorial of {number} is {result}")
````

### Multiplication Table

This code defines a function `display_multiplication_table` that prints the multiplication table for a given number up to 10. It then calls the function with an example number.

````python
def display_multiplication_table(number):
    """Displays the multiplication table for a given number."""
    print(f"Multiplication Table for {number}:")
    for i in range(1, 11):
        print(f"{number} x {i} = {number * i}")

# Example usage:
num = 7
display_multiplication_table(num)
````

### Odd or Even Number Identifier

This code defines a function `check_odd_even` that determines whether a given integer is odd or even using the modulo operator. It then calls the function with an example number and prints the result.

````python
def check_odd_even(number):
    """Checks if a given number is odd or even."""
    if number % 2 == 0:
        print(f"The number {number} is even.")
    else:
        print(f"The number {number} is odd.")

# Example usage:
num_to_check = 13
check_odd_even(num_to_check)

num_to_check = 20
check_odd_even(num_to_check)
````

### Palindrome Number Identifier

This code defines a function `is_palindrome` that checks if a given integer is a palindrome. A number is a palindrome if it reads the same forwards and backwards. The function converts the number to a string to easily reverse and compare it. It then calls the function with example numbers and prints the results.

````python
def is_palindrome(number):
    """Checks if a given number is a palindrome."""
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the original string with its reverse
    if num_str == num_str[::-1]:
        print(f"The number {number} is a palindrome.")
    else:
        print(f"The number {number} is not a palindrome.")

# Example usage:
num_to_check = 121
is_palindrome(num_to_check)

num_to_check = 123
is_palindrome(num_to_check)

num_to_check = 9009
is_palindrome(num_to_check)
````

## Implement Basic Rule Engine

### Subtask:
Implement a basic rule engine in Python that takes a student's average grade as input and applies the defined 'if-then' rules to determine if the student passes or fails the course.


**Reasoning**:
Define the Python function `decide_course_status` that takes `average_grade` as input and applies the 'if-then' rules to determine if the student passes or fails, as specified in the instructions.



````python
def decide_course_status(average_grade):
    """Determines if a student passes or fails based on their average grade."""
    if average_grade >= 60:
        return 'Passes'
    else:
        return 'Fails'

# Example usage:
print(f"Student with grade 75: {decide_course_status(75)}")
print(f"Student with grade 59: {decide_course_status(59)}")
print(f"Student with grade 60: {decide_course_status(60)}")
````
